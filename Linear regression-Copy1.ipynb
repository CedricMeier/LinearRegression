```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

def graph_linearregression_data():
    
    df=pd.read_excel('Data_Std_orsieres_50m.xlsx',usecols='B:C') #Data_Std_orsieres_50m.xlsx
    df
    x=df.data_XRF
    x
    y=df.data_labo
    y

#plt.scatter(x,y)

#machine learning Linear regression (line fitting)
    model = np.polyfit(x, y, 1)

#to know the linear equation
    model

#equation
    a=model[0]
    b=model[1]
    #d=('y','=',round(a, 2),'+',round(b,2))
#to build the line
    predict = np.poly1d(model)

#to know R-squared
    from sklearn.metrics import r2_score
    r2_score(y, predict(x))

#Data visualization

    x_lin_reg = range(0,2000)
    y_lin_reg = predict(x_lin_reg)
#plt.scatter(x, y)

     
    
    
    #plt.figure()
    plt.scatter(x, y)

    plt.plot(x_lin_reg, y_lin_reg, c='r')
     #COSMETICS
    ax =plt.gca()
    ax.set_xlabel('mg/kg de Pb(XRF)', fontsize='medium')
    ax.set_ylabel('mg/kg de Pb (laboratoire)',fontsize='medium')
    ax.set_title('Concentration en plomb XRF vs laboratoire', fontsize='medium')

    plt.tick_params(left=True,
                    bottom=True,
                    labelleft=True,
                    labelbottom=True) 
    plt.xticks(alpha=0.8)
    plt.yticks(alpha=0.8, fontsize='x-small')
    
#annotate
    ax.annotate('',xy=(1250, 1000), xytext=(1500, 500),
            arrowprops=dict(facecolor='black', shrink=0.05))
    ax.text(1500,500, f'$Y = {round(a, 2)} * x + {round(b,2)}$', family='monospace', fontsize=10)
    plt.box(False)
    #plt.show()
    #plt.savefig('fig.png')
    plt.savefig('graph_linearregression_data_v2.png',dpi=400, bbox_inches='tight')

    return plt.show()#graph

graph_linearregression_data()


```
